```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns; sns.set_style('darkgrid',{"grid.color": ".6", "grid.linestyle": ":"})
```

```python
zspec_table = pd.read_csv("ALL_Spec_Only_V7_Aper6_CCM89.csv")
```

```python
zspec_table.columns
```

```python
ra_0 = 159.17
dec_0 = -25.5244

radius = 1.5744
```

```python
def dist(ra, dec):
    return np.sqrt((ra - ra_0)**2 + (dec - dec_0)**2)
```

```python
zspec = zspec_table["z"]
g_aper_6 = zspec_table["g_aper_6"]
r_aper_6 = zspec_table["r_aper_6"]
```

```python
mask = (g_aper_6 < 40) & (r_aper_6 < 40)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

ax.scatter(g_aper_6[mask] - r_aper_6[mask], g_aper_6[mask], marker='.', color='firebrick')
ax.set_ylabel('g_aper_6', fontsize=20)
ax.set_xlabel('g_aper_6 - r_aper_6', fontsize=20)

ax.tick_params(axis='x', labelsize=20) 
ax.tick_params(axis='y', labelsize=20)
```
